this where i got the data shown below from http://lib.stat.cmu.edu/DASL/Datafiles/cigcancerdat.html

```python
import csv
```

```python
f = open('cigcancerdata.csv', 'rb')
reader = csv.reader(f)
rownum = 0
number_of_cig = []
death_from_leukemia  = []

for row in reader:
    a = row[0].split()
    b = a[1:]
    c = map(float, b)
    number_of_cig.append(c[0])
    death_from_leukemia.append(c[4])

print number_of_cig
print death_from_leukemia
    
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Smoking and cancer')
ax.set_xlabel('# of cigarettes smoked (hds per capita) ')
ax.set_ylabel('# of deaths from leukemia ( per 100 K population) ')
ax.scatter(number_of_cig, death_from_leukemia)
fig.savefig("smoking_and_cancer")

```

The graph above explains the relationship between the number of cigarettes smoked per capita and the range of deaths per 100 k population due to leukemia (directly proportional)

```python
f = open('cigcancerdata.csv', 'rb')
reader = csv.reader(f)
rownum = 0
deaths_from_lung = []
death_from_kid  = []

for row in reader:
    a = row[0].split()
    b = a[1:]
    c = map(float, b)
    deaths_from_lung.append(c[2])
    death_from_kid.append(c[3])

print deaths_from_lung
print death_from_kid
    
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Deaths per 100 K population from lung and kidney cancer ')
ax.set_xlabel('# of deaths from kidney cancer ')
ax.set_ylabel('# of deaths from lung cancer ')
ax.bar(number_of_cig, death_from_leukemia)
fig.savefig("lung_and_kidney_cancer")
```

the plot above explains the relationship between the numbers of deaths caused by lung cancer, and the ones caused by kidney cancer.. which both rely on the number of cigarettes smoked per capita.
